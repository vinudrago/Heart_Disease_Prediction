```python
X = hd.drop(columns = "target", axis = 1)
y = hd["target"]
```

```python
X
```

```python
y
```

```python
#from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 2, stratify = y)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Model Traning
# using LogisticRigression model

model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
# Model Evaluation
# Accuracy score
X_train_predication = model.predict(X_train)
traning_data_accuracy = accuracy_score(X_train_predication, y_train)
```

```python
print("Accuracy on Traning data :", traning_data_accuracy )
```

```python
X_test_predication = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predication, y_test)
print("Accuracy on Test data :", test_data_accuracy )
```
